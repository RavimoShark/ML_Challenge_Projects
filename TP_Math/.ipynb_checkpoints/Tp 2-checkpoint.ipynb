## 1 ere question

```python
import pandas as pd
import urllib
import zipfile
import sys
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt  # for plots
from matplotlib import rc
import seaborn as sns
from os import mkdir, path
from scipy import stats
from mpl_toolkits.mplot3d import Axes3D
from math import cos, sin, pi, sqrt
from statsmodels.nonparametric.kde import KDEUnivariate
from scipy.linalg import toeplitz
from numpy.linalg import eigh
from sklearn import linear_model
from sklearn import preprocessing as pr

urlGalton = 'http://josephsalmon.eu/enseignement/TELECOM/MDI720/datasets/Galton.txt'

urllib.request.urlretrieve(urlGalton, 'Galton.txt')

fieldsdata = ['Family', 'Father', 'Mother', 'Gender', 'Height', 'Kids']

DataGalton = pd.read_csv('Galton.txt', sep='\t', usecols=fieldsdata)

# Arrondis sans chiffre après la virgule

DataGalton.info()
DataGalton.Father = 2.54 * DataGalton.Father
DataGalton.Mother = 2.54 * DataGalton.Mother
DataGalton.Height = 2.54 * DataGalton.Height
DataGaltonRounded = DataGalton.round()
DataGaltonRounded.head(10)
```

# 2 eme question
Pas de données manquantes

```python
# nombre de données manquantes
print(DataGaltonRounded.isnull().sum())
```

## 3 eme question

```python
%matplotlib inline
plt.style.use('seaborn')
plt.figure(figsize=(18.5, 5.5))

# densité des tailles des pères en bleu et des mères en orange
# Je prends la moyenne pour extraire seulement un couple de parent par famille
dataFather = DataGaltonRounded.groupby(['Family']).Father.mean()
dataMother = DataGaltonRounded.groupby(['Family']).Mother.mean()

dataFather = sorted(dataFather)
dataMother = sorted(dataMother)
kdeFather = KDEUnivariate(dataFather)

# On choisit la meme bandwith
kdeFather.fit(bw=2.25, kernel='gau')
kdeMother = KDEUnivariate(dataMother)
kdeMother.fit(bw=2.25, kernel='gau')

# je choisis un interval adequate avec une granularité de 200
x_grid = np.linspace(DataGaltonRounded.Mother.min() - 30,
                     DataGaltonRounded.Father.max() + 30, 200)
FatherDensity = kdeFather.evaluate(x_grid)
MotherDensity = kdeMother.evaluate(x_grid)
plt.fill(x_grid, FatherDensity, color='blue',
         label='Father Height Density', alpha=0.7)
plt.fill(x_grid, MotherDensity, color='orange',
         label='Mother Height Density',  alpha=0.7)
plt.xlabel("Heigt in cm")
plt.ylabel("Density")
plt.legend()
plt.show()
```

# 4 eme question


```python
# Taille du père en fonction de la taille de la mère

%matplotlib inline
plt.style.use('seaborn')
plt.figure(figsize=(18.5, 10.5))

dataFather_1 = DataGaltonRounded.groupby(['Family']).Father.mean()
dataMother_1 = DataGaltonRounded.groupby(['Family']).Mother.mean()

plt.scatter(dataMother_1, dataFather_1,
            color='blue', label='Taille Père en fonction de mère')
plt.xlabel('Taille de la mère en cm')
plt.ylabel('Taille du père en cm')


# regression linéaire : X Taille mères, Y
skl_linmod = linear_model.LinearRegression(fit_intercept=True)
skl_linmod.fit(dataMother_1.values.reshape(-1, 1),
               dataFather_1.values.reshape(-1, 1))
X_to_predict = dataMother_1.values.reshape(-1, 1)

plt.plot(X_to_predict, skl_linmod.predict(X_to_predict),
         linewidth=3, label="Estimateur taille des pères", color='brown')
plt.legend(numpoints=1, loc=2)  # numpoints = 1 for nicer display
plt.xticks(np.linspace(DataGaltonRounded.Mother.min() - 10,
                       DataGaltonRounded.Mother.max() + 10, 25))
plt.yticks(np.linspace(DataGaltonRounded.Father.min() - 10,
                       DataGaltonRounded.Father.max() + 10, 25))
plt.tight_layout()
plt.legend()
plt.show()
```

Au vu du graphique aucune correlation entre la taille des pères et des mères, cela se verifie avec la droite de prediction qui scinde nos observations en deux parties un peu près égales.

## 6 eme question 


```python
DataGaltonRounded['MidParents'] = (
    DataGaltonRounded.Father + DataGaltonRounded.Mother * 1.08) / 2
DataGaltonRounded.head()
print(DataGaltonRounded.Height.min())
print(DataGaltonRounded.Height.max())
print(DataGaltonRounded.MidParents.min())
print(DataGaltonRounded.MidParents.max())
```

## 7 eme question 

```python
# Estimateur Theta0 et Theta1 par les moindres carrés avec intercept à false
skl_linmod1 = linear_model.LinearRegression(fit_intercept=True)
skl_linmod1.fit(DataGaltonRounded.MidParents.values.reshape(-1, 1),
                DataGaltonRounded.Height.values.reshape(-1, 1))
Theta1 = skl_linmod1.coef_
Theta0 = skl_linmod1.intercept_
Theta1Chap = float(Theta1[0][0])
Theta0Chap = float(Theta0[0])
print("Theta1Chap = ", Theta1Chap)
print("Theta0Chap = ", Theta0Chap)
# utlisation des dataframes pour le calcul théorique
DataGaltonRounded['EMPM'] = (DataGaltonRounded.MidParents - DataGaltonRounded.MidParents.mean()
                             ) * (DataGaltonRounded.Height - DataGaltonRounded.Height.mean())
DataGaltonRounded['VPM'] = (
    DataGaltonRounded.MidParents - DataGaltonRounded.MidParents.mean())**2
Theta1theo = DataGaltonRounded.EMPM.sum() / DataGaltonRounded.VPM.sum()
Theta0theo = DataGaltonRounded.Height.mean() - Theta1theo * \
    DataGaltonRounded.MidParents.mean()
print("le Theta théorique 0 est : ", Theta0theo)
print("le Theta 1 théorique est: ", Theta1theo)
print("Verification numérique  ", np.isclose(Theta1Chap, Theta1theo))
print("Verification numérique  ", np.isclose(Theta0Chap, Theta0theo))
```

## 8 eme question

```python
# Valeurs à predires pour les filles,
DaughterToPredict = DataGaltonRounded[DataGaltonRounded['Gender']
                                      == 'F'].MidParents.values.reshape(-1, 1)

# Valeurs à predires pour les garcons
BoysToPredict = DataGaltonRounded[DataGaltonRounded['Gender']
                                  == 'M'].MidParents.values.reshape(-1, 1)

# Valeurs Résultats pour les Filles
DaughterRes = DataGaltonRounded[DataGaltonRounded['Gender'] == 'F'].Height

# Valeurs Résultats pour les garcons
BoyRes = DataGaltonRounded[DataGaltonRounded['Gender'] == 'M'].Height

# Graph
x_grid = np.linspace(DataGaltonRounded.Height.min() - 30,
                     DataGaltonRounded.Height.max() + 30, )


fig = plt.figure(figsize=(18.5, 10.5))
plt.scatter(DaughterToPredict, skl_linmod1.predict(DaughterToPredict),
            label="Estimateur des moindre carres pour les filles", color='turquoise', alpha=0.7)
plt.scatter(BoysToPredict, skl_linmod1.predict(BoysToPredict),
            label="Estimateur des moindre carres pour les garcons", color='red', alpha=1)
plt.scatter(DataGaltonRounded[DataGaltonRounded['Gender'] == 'F'].MidParents, DaughterRes,
            label='observations pour les filles', color='orange')
plt.scatter(DataGaltonRounded[DataGaltonRounded['Gender'] == 'M'].MidParents, BoyRes,
            label='observations pour les garcons', color='blue')
plt.title("Moindre Carré vs Observations pour Filles et Garcons")
plt.xticks(np.linspace(155, 198, 20))
plt.yticks(np.linspace(120, 210, 20))
plt.legend()
plt.show
```

## 9 eme Question

```python
fig = plt.figure(figsize=(18.5, 10.5))

# Estimateur de la densité des résidus pour l'échantillon total
residualChildMidParent = DataGaltonRounded.Height.values.reshape(-1, 1)
- skl_linmod1.predict(DataGaltonRounded.MidParents.values.reshape(-1, 1))
residualChildMidParent = residualChildMidParent - Theta0Chap
dataResidual = sorted(residualChildMidParent)
kdeResidual = KDEUnivariate(dataResidual)
kdeResidual.fit(bw=2, kernel='gau')

# Proportion alphaG et alphaF de l'échantillon :

AlphaF = DataGaltonRounded[DataGaltonRounded.Gender ==
                           'F'].count() / DataGaltonRounded.Gender.count()

# A cette époque les Transexuels ne sont pas recencés donc:
AlphaG = 1 - AlphaF


# Résidus pour les filles et garcons
residualFille = DaughterRes
- skl_linmod1.predict(DaughterToPredict)

residualGarcon = BoyRes
- skl_linmod1.predict(BoysToPredict)

residualFille = residualFille - Theta0Chap
residualGarcon = residualGarcon - Theta0Chap

# densité

dataResidualFille = sorted(residualFille)
kdeResidualFille = KDEUnivariate(dataResidualFille)
kdeResidualFille.fit(bw=2, kernel='gau')
dataResidualgarcon = sorted(residualGarcon)
kdeResidualGarcon = KDEUnivariate(dataResidualgarcon)
kdeResidualGarcon.fit(bw=2, kernel='gau')

# Graphique
x_grid = np.linspace(50, 175, 350)
ResidualDensity = kdeResidual.evaluate(x_grid)
ResidualDensityFille = kdeResidualFille.evaluate(x_grid)
ResidualDensityGarcon = kdeResidualGarcon.evaluate(x_grid)

plt.fill(x_grid, ResidualDensity, color='blue',
         label='Densité des résidus echantillon total', alpha=0.45)
plt.fill(x_grid, ResidualDensityFille * AlphaF.Kids, color='pink',
         label='Densité des résidus des filles', alpha=0.7)
plt.fill(x_grid, ResidualDensityGarcon * AlphaG.Kids, color='brown',
         label='Densité des résidus des garcons', alpha=0.7)

plt.xlabel("Error in cm")
plt.ylabel("Density")
plt.legend()
plt.show()
```

L'hypothèse de distribution des résidus selon la loi normale ne parait  pas satisfaisante pour l'échantillon fille garçon réuni, en effet on voit deux bosses qui pourraient correspondre à la densité des résidus des filles et des garcons. Nous le verifions visuellement avec deux belles gaussiennes centrées sur 105 et 120.


## 10 eme question 

```python
# regression des xi sur les yi

skl_linmod2 = linear_model.LinearRegression(fit_intercept=True)
skl_linmod2.fit(DataGaltonRounded.Height.values.reshape(-1, 1),
                DataGaltonRounded.MidParents.values.reshape(-1, 1))

Alpha0Chap = float(skl_linmod2.intercept_[0])
Alpha1Chap = float(skl_linmod2.coef_[0][0])

# Verifiation numérique

ymean = DataGaltonRounded.Height.mean()
xmean = DataGaltonRounded.MidParents.mean()
yvar = DataGaltonRounded.Height.var()
xvar = DataGaltonRounded.MidParents.var()
print(xmean, ymean, xvar, yvar)
Alpha0Theo = xmean + (ymean / xmean) * (xvar / yvar) * (Theta0Chap - ymean)
Alpha1Theo = xvar / yvar * Theta1Chap
print("Alpha0Chap = ", Alpha0Chap)
print("Alpha1Chap = ", Alpha1Chap)
print("Alpha0Theo = ", Alpha0Theo)
print("Alpha1Theo = ", Alpha1Theo)
print("Les AlphasO sont ils proches", np.isclose(Alpha0Theo, Alpha0Chap))
print("Les Alphas1 sont ils proches ", np.isclose(Alpha1Theo, Alpha1Chap))
```

## 13 eme question

```python
from sklearn import preprocessing


def GetKey(x):
    return abs(x[0])


# Centré et reduire
AutoSubset = DataAuto.iloc[::, :7:]
AutoSubset.iloc[::, ::] = preprocessing.scale(AutoSubset.iloc[::, ::])

X = AutoSubset.iloc[::, 1::]
Y = AutoSubset.mpg
# Regresssion linéaire
skl_linmod4 = linear_model.LinearRegression(fit_intercept=True)
skl_linmod4.fit(AutoSubset.iloc[::, 1::], AutoSubset.mpg)


# Estimateurs

Theta0AutoChapCr = skl_linmod4.intercept_
Theta1AutoChapCr = skl_linmod4.coef_
Res_1 = list(map(lambda x, y: (x, y), Theta1AutoChapCr,
                 AutoSubset.columns.values[1::]))
Res_1 = sorted(Res_1, key=GetKey, reverse=True)
print("l'estimteur Theta0 sur le subset total est \n", Theta0AutoChapCr)
print("l'estimteur Theta1 sur le subset total est \n", Res_1)
```

On peut donc remarquer que les deux variables les plus signifiatives influent sur la consommation sont:
    - le poids est le facteur le plus determinant
    - model year donc l'année de fabrication de la voiture peut s'expliquer par une réduction de la cylindrée 
    et/ou une meilleure efficacité du rendement du moteur. Grace aux remplacements
    des carburateurs par l'injection directe d'essence dans la chambre de combustion.
De plus du fait d'avoir centré et réduit nos features nous avons réduit le Theta à pratiquement 0.

## 14 eme question

```python
# calcul des residus
Autoresidual = AutoSubset.mpg - (skl_linmod4.predict(AutoSubset.iloc[::, 1::]))

# norme produit scalaire du vecteur par lui meme

Norm = np.dot(Autoresidual.T, Autoresidual)
NormUni = Norm / (len(Autoresidual) - len(Theta1AutoChapCr))
print('La norme au carré est : ', Norm)
print('La norme au carré unitaire est : ', NormUni)

# Verification par le calcul
DiffObsMoy = AutoSubset.mpg - AutoSubset.mpg.mean()
DiffObsMoyNorm = np.dot((DiffObsMoy).T, DiffObsMoy)
DiffPredMoy = skl_linmod4.predict(
    AutoSubset.iloc[::, 1::]) - AutoSubset.mpg.mean()
DiffPredMoyNorm = np.dot(DiffPredMoy.T, DiffPredMoy)
print('Verification numérique: ', np.isclose(
    DiffObsMoyNorm, Norm + DiffPredMoyNorm))
```

## 16 eme question

```python
# Polynomial features
poly = pr.PolynomialFeatures(2, interaction_only=False)
# On Applique aux données brutes
DataPoly = poly.fit_transform(DataAuto.iloc[::, 1:7:])


# On recentre et reduit on ne rescale pas la colonne des constantes
DataPoly = pr.scale(DataPoly[::, 1::])

# On applique ensuite le modéle
X = DataPoly
Y = DataAuto.mpg

skl_linmod7 = linear_model.LinearRegression(fit_intercept=True)
skl_linmod7.fit(X, Y)
# Estimateurs

Theta0AutoPolyChapCr = skl_linmod7.intercept_
Theta1AutoPolyChapCr = list(skl_linmod7.coef_)
FieldsValue = poly.get_feature_names(DataAuto.iloc[::, 1:7:].columns.values)

# Result as tuples
Res = list(map(lambda x, y: (x, y), Theta1AutoPolyChapCr, FieldsValue[1::]))
SortedRes = sorted(Res, key=GetKey, reverse=True)
print('la variable la plus significative est: ', SortedRes[0][1])
print("A titre d'information voici les 5 variables les plus significatives avec leurs poids",
      SortedRes[:5:])
```

```python
# gérer la variable origin

OrDumm = pd.get_dummies(DataAuto['origin'])
DataAutoOrDumm = pd.concat([DataAuto, OrDumm], axis=1)
DataAutoOrDumm.rename(
    columns={1.0: 'usa', 2.0: 'europe', 3.0: 'japan'}, inplace=True)

# on fait tourner le modele et on drop la colonne usa qui devient notre référence
Xi = DataAutoOrDumm.drop(['mpg', 'car name', 'origin', 'usa'], axis=1)

XAuto = pr.scale(Xi)
YAuto = pr.scale(DataAuto.mpg)

skl_linmod8 = linear_model.LinearRegression(fit_intercept=False)
skl_linmod8.fit(XAuto, YAuto)

Theta1wOr = skl_linmod8.coef_
FieldsName = list(Xi.columns.values)
ResOr = list(map(lambda x, y: (x, y), Theta1wOr, FieldsName))
SortedResOr = sorted(ResOr, key=GetKey, reverse=True)
print("Voici les variables avec leurs poids",
      SortedResOr)
```

# 19 eme question

```python
# Svd Matrice X à la 13 eme question
X = pr.scale(DataAuto.iloc[::, 1:7:])
U, s, V = np.linalg.svd(X, full_matrices=False)

# démontrons que H est un projeté orthogonal: H**2=H
H = np.dot(U, U.T)
H2 = np.dot(H.T, H)
print("H = H**2: ", np.all(np.isclose(H, H2)))
print("H = H.T: ", np.all(np.isclose(H, H.T)))
print("H est bien un projeté orthogonal")
```
